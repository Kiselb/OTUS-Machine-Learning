```python
import datetime

def datetime_build(year, month, day, time):
    if int(time) >= 2400:
        return datetime.timedelta(days=1) + datetime.datetime(year, month, day, (int(time) - 2400) // 100,  int(time) % 100, 0, 0)
    else:
        return datetime.datetime(year, month, day, int(time) // 100,  int(time) % 100, 0, 0)
```

### Часть 1. Пишем симуляцию.

В предложенной в задании реализации функции симуляции заражения, ничего не сказано об использовании времени отправления и прибытия рейса. Но время прибытия и отправления из аэропорта является ключевым при рассчёте распространения инфекции. Поэтому я реализовал свой алгоритм симуляции. Работа алгоритма проста: для каждой минуты рассчитывается заражённость аэропорта, если было прибытие из заражённого аэропоата (аэропортов). Конечно, представленный алгоритм не является оптимальным, понятно, что можно оптимизировать. Я не стал производить оптимизацию, чтобы сохранить наглядность реализации

```python
import random
from array import *
```

```python
def infection_propagation(infecting_probably):    
    infected_airports = set()
    infected_airports.add('ABE')
    infecting_by_hours = array('i', [])
    
    for current_minute in range(1, int((max_simulation_time - min_simulation_time).total_seconds() / 60) + 1):
        current_arrivals = data.loc[data['MinuteTick'] == current_minute]
        if current_arrivals.shape[0] > 0:
            for index, row in current_arrivals.iterrows():
                if current_arrivals.at[index, 'Origin'] in infected_airports:
                    if random.uniform(0, 1) >= (1 - infecting_probably):
                        infected_airports.add(current_arrivals.at[index, 'Dest'])
        if current_minute % 60 == 0:
            infecting_by_hours.append(len(infected_airports))
    return np.array(infecting_by_hours)
```

### Часть 2. Как инфекционность (вероятность заражения) болезни влияет на скорость распространения?

```python
for i in range(1, 11):
    result = infection_propagation(0.01)
    if i == 1:
        simulation001 = np.empty((0, len(result)))
    simulation001 = np.append(simulation001, result.reshape(1, len(result)), axis=0)

    result = infection_propagation(0.05)
    if i == 1:
        simulation005 = np.empty((0, len(result)))
    simulation005 = np.append(simulation005, result.reshape(1, len(result)), axis=0)
    
    result = infection_propagation(0.10)
    if i == 1:
        simulation010 = np.empty((0, len(result)))
    simulation010 = np.append(simulation010, result.reshape(1, len(result)), axis=0)
    
    result = infection_propagation(0.50)
    if i == 1:
        simulation050 = np.empty((0, len(result)))
    simulation050 = np.append(simulation050, result.reshape(1, len(result)), axis=0)
    
    result = infection_propagation(1.00)
    if i == 1:
        simulation100 = np.empty((0, len(result)))
    simulation100 = np.append(simulation100, result.reshape(1, len(result)), axis=0)
```

```python
plt.plot(np.arange(0, simulation001.shape[1]), simulation001.mean(axis=0), label="0.01")
plt.plot(np.arange(0, simulation005.shape[1]), simulation005.mean(axis=0), label="0.05")
plt.plot(np.arange(0, simulation010.shape[1]), simulation010.mean(axis=0), label="0.10")
plt.plot(np.arange(0, simulation050.shape[1]), simulation050.mean(axis=0), label="0.50")
plt.plot(np.arange(0, simulation100.shape[1]), simulation100.mean(axis=0), label="1.00")
plt.legend()
plt.xlabel("Hours")
plt.ylabel("Airports Quantity")
plt.title("Infecting Simulation")
```

### Часть 3. Может ли информация о сети помочь предсказать, кто будет заражен?

```python
def build_flatten_graph():
    result = dict()
    for row in data.itertuples():
        if row.Origin > row.Dest:
            key = row.Dest + row.Origin
        else:
            key = row.Origin + row.Dest
        if key in result:
            result[key] += 1
        else:
            result[key] = 1
    return result
```

```python
flatten_graph = build_flatten_graph()
```

```python
len(flatten_graph)
```

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
for key in flatten_graph.keys():
    G.add_edge(key[:3], key[3:], object=flatten_graph[key] / data.shape[0])
```

```python
G.number_of_nodes()
```

```python
options = {
    'node_color': 'black',
    'edge_color': 'lightgray',
    'node_size': 10,
    'width': 1,
}
plt.figure(figsize=(7, 7))
nx.draw_circular(G, **options)
```

#### Вычисление медианного времени до заражения каждого города

```python
infected_airports_time = dict()
infected_airports = set()
infected_airports.add('ABE')

for i in range(1, 51):
    for current_minute in range(1, int((max_simulation_time - min_simulation_time).total_seconds() / 60) + 1):
        current_arrivals = data.loc[data['MinuteTick'] == current_minute]
        if current_arrivals.shape[0] > 0:
            for index, row in current_arrivals.iterrows():
                if current_arrivals.at[index, 'Origin'] in infected_airports:
                    if random.uniform(0, 1) >= (1 - 0.5):
                        infected_airports.add(current_arrivals.at[index, 'Dest'])
                        if current_arrivals.at[index, 'Dest'] not in infected_airports_time:
                            infected_airports_time[current_arrivals.at[index, 'Dest']] = []
                        if (len(infected_airports_time[current_arrivals.at[index, 'Dest']]) != i):                        
                            infected_airports_time[current_arrivals.at[index, 'Dest']].append(current_minute)
```

#### Вычисление метрик графа

```python
infected_airports_time_median = dict()
for key in infected_airports_time.keys():
    infected_airports_time_median[key] = np.median(infected_airports_time[key])
```

```python
infected_airports_time_median_keys = list(infected_airports_time_median.keys())
infected_airports_time_median_keys.sort()
infected_airports_time_median_sorted = [infected_airports_time_median[key] for key in infected_airports_time_median_keys]
```

#### Коэффициент корреляции Спирмана

```python
from scipy.stats import spearmanr
```

```python
corr, pval = spearmanr(clustering_items_sorted, infected_airports_time_median_sorted)
print("Clustering Spearman's correlation coefficient:", corr)
print("p-value:", pval)
```

```python
corr, pval = spearmanr(graph_degree_sorted, infected_airports_time_median_sorted)
print("Degree Spearman's correlation coefficient:", corr)
print("p-value:", pval)
```

```python
corr, pval = spearmanr(bc_sorted, infected_airports_time_median_sorted)
print("Betweenness Centrality Spearman's correlation coefficient:", corr)
print("p-value:", pval)
```

#### Выводы

Корреляция между Clustering и медианным временем заражения незначительна. Поскольку коэффициент описывает топологию графа, а граф рейсов очень насыщен, то влияние топологии незначительно. 

Корреляция между Degree и медианным временем заражения значима, но в обратную сторону. Корреляция интепретируема: чем с большими узлами связан данный узел, тем меньше время, необходимое для его заражения.

Корреляция между Betweenness Centrality и медианным временем заражения весьма значительна, но в обратную сторону. Корреляция интепретируема: чем значимее узел, значимый в том смысле, что через него проходит значительное количество кратчайших путе графа, тем меньше время, необходимое для его заражения.
