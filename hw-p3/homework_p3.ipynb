#### 3.1. Статистический метод - три сигмы

```python
up_bound = data_scaled.mean(axis=0) + 3 * data_scaled.std(axis=0)
low_bound = data_scaled.mean(axis=0) - 3 * data_scaled.std(axis=0)
sigma3 = np.where((data_scaled < low_bound) | (data_scaled > up_bound), 1, 0).any(axis=1).astype('int8')
sigma3
```

```python
from operator import * 
```

```python
outlier_n = countOf(sigma3, 1)
outlier_p = countOf(sigma3, 1) / (countOf(sigma3, 0) + countOf(sigma3, 1))    

print('Количество выбросов в выборке для 3 сигм: {}'.format(outlier_n))
print('Процент выбросов в выборке для 3 сигм: {:.2%}'.format(outlier_p))
```

#### 3.2. Статистический метод - IQR

```python
data_anomalies = pd.DataFrame(columns = data.columns)
threshold=1.5
for column in data.columns:
    IQR = data[column].quantile(0.75) - data[column].quantile(0.25)
    ubound = data[column].quantile(0.75) + (IQR * threshold)
    lbound = data[column].quantile(0.25) - (IQR * threshold)
    data_anomalies[column] = pd.concat([data[column] > ubound, data[column] < lbound], axis=1).sum(axis=1) >= 1
```

```python
data_anomalies.sum(axis=1) > 0
```

```python
print('Количество выбросов в выборке: {}'.format(sum(data_anomalies.sum(axis=1) > 0)))
print('Процент выбросов в выборке: {:.2%}'.format(sum(data_anomalies.sum(axis=1) > 0) / len(data)))
```

```python
from sklearn.base import BaseEstimator
from scipy.spatial.distance import cdist
```

```python
class DistanceOutliers(BaseEstimator):
    """
    Parameters:
    - metric: string, default - euclidean
        metric to use for distance calculation (see scipy.spatial.distance.cdist)
    - percentile: float in range [0, 100]
        hyperparameter which sets the threshold for anomalies
    """
    def __init__(self, metric='euclidean', percentile=90):
        self.metric = metric
        self.percentile = percentile

    def fit(self, X):
        self.centroid = np.mean(X, axis=0).values.reshape(-1, 1).T
        distances_train = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        self.threshold = np.percentile(distances_train, self.percentile)

    def predict(self, X):
        distances = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        predictions = (distances > self.threshold).astype(int)
        return predictions
```

```python
euclidian_model = DistanceOutliers(metric='euclidean', percentile=95)
euclidian_model.fit(data_scaled)
euclidian_outliers = euclidian_model.predict(data_scaled)
```

```python
print('Количество выбросов в выборке: {}'.format(sum(euclidian_outliers)))
print('Процент выбросов в выборке: {:.2%}'.format(sum(euclidian_outliers)/len(data)))
```
