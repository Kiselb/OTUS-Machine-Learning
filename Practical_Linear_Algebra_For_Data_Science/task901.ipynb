```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = np.random.randn(10, 10)
print(matrix)
```

```python
Q, R = np.linalg.qr(matrix, "complete")
```

```python
error = (Q.T @ Q) - np.eye(10)
print(np.sqrt(np.sum(error **2)))
print(Q.T @ Q)
```

```python
error = (Q @ Q.T) - np.eye(10)
print(np.sqrt(np.sum(error **2)))
print(Q @ Q.T)
```

```python
error = (np.linalg.inv(Q) @ Q) - np.eye(10)
print(np.sqrt(np.sum(error **2)))
print(np.linalg.inv(Q) @ Q)
```

```python
error = (Q @ np.linalg.inv(Q)) - np.eye(10)
print(np.sqrt(np.sum(error **2)))
print(Q @ np.linalg.inv(Q))
```
