```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
import sympy as sym
```

```python
rank = 6
M = np.random.randn(rank, rank) 
```

```python
T = np.eye(rank)
```

```python
R1 = np.zeros_like(M)

for coli in range(rank):
  R1[:,coli] = np.linalg.inv(M.T @ M) @ M.T @ T[:,coli]
```

```python
R2 = np.linalg.inv(M.T @ M) @ M.T
```

```python
R3 = np.linalg.inv(M)
```

```python
_,axs = plt.subplots(1, 3, figsize = (10, 6))

axs[0].imshow(R1 @ M, cmap = 'gray')
axs[0].set_title('Via column-wise LS')

axs[1].imshow(R2 @ M, cmap = 'gray' )
axs[1].set_title('Via matrix-wise LS')

axs[2].imshow(R3 @ M, cmap = 'gray' )
axs[2].set_title('Via inv()')

for a in axs: a.set(xticks = [], yticks = [])

plt.tight_layout()
```
