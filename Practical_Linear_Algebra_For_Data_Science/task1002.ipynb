```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
def matrix_construct_rank(rows, cols, rank):
    return np.random.randn(rows, rank) @ np.random.randn(rank, cols)
```

```python
matrix = matrix_construct_rank(6, 8, 3)
```

```python
P, L, U = scipy.linalg.lu(matrix)
```

```python
_,axs = plt.subplots(1,3,figsize=(12,7))

axs[0].imshow(matrix, vmin = -1, vmax = 1, cmap = 'gray')
axs[0].set_title(f'Matrix, rank={np.linalg.matrix_rank(matrix)}')

axs[1].imshow(L, vmin = -1, vmax = 1, cmap = 'gray')
axs[1].set_title(f'L, rank={np.linalg.matrix_rank(L)}')

axs[2].imshow(U, vmin = -1, vmax = 1, cmap = 'gray')
axs[2].set_title(f'U, rank={np.linalg.matrix_rank(U)}')

plt.tight_layout()
```
