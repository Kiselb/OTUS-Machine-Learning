```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rank = 42
rnum = 123
VX = np.zeros((rnum, rank), dtype='complex')
```

```python
for i in range(rnum):
    M = np.random.randn(rank, rank)
    VX[i,:]= np.linalg.eig(M)[0] / np.sqrt(rank)
```

```python
plt.figure(figsize=(6,6))

plt.plot(np.real(VX),np.imag(VX),'o',markerfacecolor='white')
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])
plt.xlabel('Real')
plt.ylabel('Imag')
```
