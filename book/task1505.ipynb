```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import pandas as pd
import seaborn as sns
```

```python
rank = 200
S1 = np.random.randn(rank, 2)
```

```python
S1[:,1] = S1[:,0] + S1[:,1]
S1 = S1 + np.array([2, -1])
```

```python
S2 = np.random.randn(rank, 2)
S2[:,1] = S2[:,0] + S2[:,1]
```

```python
data = np.vstack((S1, S2))
labels  = np.append(np.zeros(rank),np.ones(rank))
```

```python
ax = sns.jointplot(x = data[:,0], y = data[:,1], hue = labels)
ax.ax_joint.set_xlabel('Data axis 1')
ax.ax_joint.set_ylabel('Data axis 2')
ax.plot_joint(sns.kdeplot)
```
