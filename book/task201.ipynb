```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v+w
vMinusW = v-w
```

```python
print(v)
print(w)
print(vPlusW)
print(vMinusW)
```

```python
plt.plot(0,0,'ko',markerfacecolor='k',markersize=7)

# add axis lines
plt.plot([-6,6],[0,0],'--',color=[.8,.8,.8],zorder=-1)
plt.plot([0,0],[-6,6],'--',color=[.8,.8,.8],zorder=-1)

# make the plot look nicer
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')

plt.arrow(0,0,v[0],v[1],head_width=.5,width=.1, length_includes_head=True)
plt.arrow(v[0],v[1],w[0],w[1],head_width=.5,width=.1, length_includes_head=True)
plt.arrow(0,0,vPlusW[0],vPlusW[1],head_width=.5,width=.1, length_includes_head=True)

plt.show()
```

```python
plt.plot(0,0,'ko',markerfacecolor='k',markersize=7)

# add axis lines
plt.plot([-6,6],[0,0],'--',color=[.8,.8,.8],zorder=-1)
plt.plot([0,0],[-6,6],'--',color=[.8,.8,.8],zorder=-1)

# make the plot look nicer
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')

plt.arrow(0,0,v[0],v[1],head_width=.5,width=.1, length_includes_head=True)
plt.arrow(0,0,w[0],w[1],head_width=.5,width=.1, length_includes_head=True)
plt.arrow(w[0],w[1],vMinusW[0],vMinusW[1],head_width=.5,width=.1, length_includes_head=True)

plt.show()
```
