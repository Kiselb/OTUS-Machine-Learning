```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import pandas as pd
import seaborn as sns
```

```python
A = np.hstack((np.random.randn(1000, 1), .05*np.random.randn(1000, 1)))
```

```python
theta = -np.pi / 6
R1 = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta),  np.cos(theta)]])
```

```python
theta = -np.pi / 3
R2 = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta),  np.cos(theta)]])
```

```python
M = np.vstack((A @ R1, A @ R2))
M.shape
```

```python
U, S, VT = np.linalg.svd(M - np.mean(M, axis = 0, keepdims = True))
```

```python
# not necessary: convert singular values into eigenvalues
S = S **2 / (M.shape[0] - 1)
```

```python
# also not necessary: up-scale the singular vectors for visualization
VT *= 2
```

```python
plt.figure(figsize=(7,7))

# the data
plt.plot(M[:,0], M[:,1], 'ko', markerfacecolor='w')

# eigenvectors
plt.plot([0, VT[0,0]], [0, VT[1,0]], 'r--', linewidth = 5, label = 'Comp 1')
plt.plot([0, VT[0,1]], [0, VT[1,1]], 'r:', linewidth = 5, label = 'Comp 2')

plt.legend()
plt.grid()
```
