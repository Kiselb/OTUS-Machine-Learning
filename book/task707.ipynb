```python
import numpy as np
import matplotlib.pyplot as plt

# setup animation
import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')


# to read an image from a url (io) and convert it to grayscale (color)
from skimage import io,color
# convolution
from scipy.signal import convolve2d
```

```python
bathtub = io.imread('https://upload.wikimedia.org/wikipedia/commons/6/61/De_nieuwe_vleugel_van_het_Stedelijk_Museum_Amsterdam.jpg')
smooth_bathtub = np.zeros(bathtub.shape)
```

```python
vkernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
hkernel = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
])
```

```python
bathtub2d = color.rgb2gray(bathtub)
```

```python
vsmooth = convolve2d(bathtub2d, vkernel, mode='same')
hsmooth = convolve2d(bathtub2d, hkernel, mode='same')
```

```python
fig,ax = plt.subplots(2,2,figsize=(16,8))

ax[0,0].imshow(vkernel,cmap='gray')
ax[0,0].set_title('Vertical kernel')
ax[0,0].set_yticks(range(3))

ax[0,1].imshow(hkernel,cmap='gray')
ax[0,1].set_title('Horizontal kernel')
ax[0,1].set_yticks(range(3))

ax[1,0].imshow(vsmooth,cmap='gray',vmin=0,vmax=.01)
ax[1,0].axis('off')

ax[1,1].imshow(hsmooth,cmap='gray',vmin=0,vmax=.01)
ax[1,1].axis('off')
```
