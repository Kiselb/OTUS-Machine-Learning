```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
kernel_1 = np.array([0,.1,.3,.8,1,.8,.3,.1,0])
kernel_1 = kernel_1 / np.sum(kernel_1)
print(kernel_1)
```

```python
kernel_2 = np.array([0,.1,.3,.8,-1,.8,.3,.1,0])
kernel_2 = kernel_2 / np.sum(kernel_2)
print(kernel_2)
```

```python
signal_length = 100
signal = np.random.randn(signal_length)
```

```python
filter_1 = np.zeros(signal_length)
filter_2 = np.zeros(signal_length)
```

```python
for i in range(len(signal) - len(kernel_1) + 1):
    filter_1[i] = np.dot(kernel_1, signal[i: i + len(kernel_1)])
for i in range(len(signal) - len(kernel_2) + 1):
    filter_2[i] = np.dot(kernel_2, signal[i: i + len(kernel_2)])
```

```python
_,axs = plt.subplots(3,2,figsize=(12,4))
axs[0][0].plot(kernel_1,'ks-')
axs[0][0].set_title('Kernel')
axs[0][0].set_xlim([-1, signal_length])

axs[0][1].plot(signal, color='k', label='Original', linewidth=1)
axs[0][1].legend()

axs[1][0].plot(signal, color='k', label='Original', linewidth=1)
axs[1][0].plot(filter_1,'--',color=[.6,.6,.6],label='Smoothed',linewidth=2)
axs[1][0].legend()

axs[1][1].plot(filter_1,'--',color=[.6,.6,.6],label='Smoothed',linewidth=2)
axs[1][1].legend()

axs[2][0].plot(signal, color='k', label='Original', linewidth=1)
axs[2][0].plot(filter_2,'--',color=[.6,.6,.6],label='Smoothed',linewidth=2)
axs[2][0].legend()

axs[2][1].plot(filter_2,'--',color=[.6,.6,.6],label='Smoothed',linewidth=2)
axs[2][1].legend()
```
