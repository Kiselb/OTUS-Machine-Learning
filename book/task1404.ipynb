```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
rank_1 = 7
rank_2 = 4
U = np.random.randn(rank_1, rank_1)
V = np.random.randn(rank_2, rank_2)
```

```python
S = np.zeros((rank_1, rank_2))
S[0, 0] = 100
S[1, 1] = 50
S[2, 2] = 25
S[3, 3] = 1
S
```

```python
M = U @ S @ V
M
```

```python
_,axs = plt.subplots(1,4,figsize=(12,6))

axs[0].imshow(M, aspect='equal', cmap='gray')
axs[0].set_title(f'M (cond={np.linalg.cond(M):.3f})')

axs[1].imshow(U, aspect='equal', cmap='gray')
axs[1].set_title(f'U (cond={np.linalg.cond(U):.3f})')

axs[2].imshow(S, aspect='equal', cmap='gray')
axs[2].set_title(f'S (cond={np.linalg.cond(S):.3f})')

axs[3].imshow(V, aspect='equal', cmap='gray')
axs[3].set_title(f'V (cond={np.linalg.cond(V):.3f})')
```
