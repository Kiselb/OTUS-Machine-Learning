```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
from scipy.linalg import toeplitz
```

```python
scalars = np.linspace(0, 50, 40) # range of scaling parameters (0 to 50 in 40 steps)
experiments = 10
norms_matrix = np.zeros((len(scalars), experiments))
```

```python
for scalar in range(len(scalars)):
    for experiment in range(experiments):
        matrix = np.random.randn(10, 10) * scalars[scalar]
        norms_matrix[scalar, experiment] = np.linalg.norm(matrix, 'fro')
```

```python
plt.plot(scalars, np.mean(norms_matrix, axis=1), 'ko-')
plt.xlabel('Matrix scalar')
plt.ylabel('Matrix Frobenius norm')
plt.show()
```
