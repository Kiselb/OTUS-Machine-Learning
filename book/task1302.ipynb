```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.array([[-1,1], [-1,2]])
```

```python
L, V = np.linalg.eig(M)
```

```python
NV = np.random.randn(2,2)
```

```python
MV = M @ V
MNV = M @ NV
```

```python
_,axs = plt.subplots(1,2,figsize=(10,6))

# the two eigenvectors
axs[0].plot([0,V[0,0]],[0,V[0,1]],'k',linewidth=2,label='V1')
axs[0].plot([0,MV[0,0]],[0,MV[0,1]],'k--',linewidth=2,label='MV1')

axs[0].plot([0,V[1,0]],[0,V[1,1]],'r',linewidth=2,label='V2')
axs[0].plot([0,MV[1,0]],[0,MV[1,1]],'r--',linewidth=2,label='MV2')

# the two non-eigenvectors
axs[1].plot([0,NV[0,0]],[0, NV[0,1]],'k',linewidth=2,label='NV1')
axs[1].plot([0,MNV[0,0]],[0,MNV[0,1]],'k--',linewidth=2,label='MNV1')

axs[1].plot([0,NV[1,0]],[0,NV[1,1]],'r',linewidth=2,label='NV2')
axs[1].plot([0,MNV[1,0]],[0,MNV[1,1]],'r--',linewidth=2,label='MNV2')


# adjust the graphs a bit
for i in range(2):
  axs[i].axis('square')
  axs[i].set_xlim([-1.5,1.5])
  axs[i].set_ylim([-1.5,1.5])
  axs[i].grid()
  axs[i].legend()

plt.show()
```
