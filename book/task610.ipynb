```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
from scipy.linalg import toeplitz
```

```python
size_sequence = np.arange(3, 31)
iterations = 100
determinants = np.zeros((len(size_sequence), iterations))
```

```python
for j in range(len(size_sequence)):
  for i in range(iterations):
      
    # step 1
    matrix = np.random.randn(size_sequence[j], size_sequence[j])
    
    # step 2
    matrix[:,0] = matrix[:,1]
    
    # step 3
    determinants[j,i] = np.abs(np.linalg.det(matrix))
```

```python
plt.figure(figsize=(6,4))
plt.plot(size_sequence **2, np.log(np.mean(determinants, axis=1)),'ks-', linewidth=3)
plt.xlabel('Number of elements in the matrix')
plt.ylabel('Log determinant')
plt.title('Empirical determinants of singular matrices')
```
